```python
import pandas as pd
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('data.csv')
```

```python
df1
```

```python
df1 = df1[~df1["cid"].isnull()]
```

```python
df1["cid"] = df1[["cid"]].astype(int)
```

```python
del df1['Label']
```

```python
data2 = pd.read_csv('ds.csv')
```

```python
df1.rename(columns={'cid':'senyawa'}, inplace=True)
df1.rename(columns={'Nama.Protein':'protein'}, inplace=True)
```

```python
data2.rename(columns={'cid':'senyawa'}, inplace=True)
data2.rename(columns={'Nama.Protein':'protein'}, inplace=True)
```

```python
data2
```

```python
del data2['Unnamed: 0']
data2
```

```python
label = []
for _, row in data2.iterrows():
    senyawa = row["senyawa"]
    protein = row["protein"]

    cek_ada = (df1[["senyawa", "protein"]].values ==
               [senyawa, protein]).all(axis=1).any()
    if cek_ada:
        label.append(1)
    else:
        label.append(0)

data2['label'] = label
```

```python
print(data2)
```

```python
data2.to_csv('label.csv', index=False)
```
